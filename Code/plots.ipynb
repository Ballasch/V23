```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import os
import matplotlib.colors as col
```

# Erstens Vorbereitung

```python
sns.set()
cmap = sns.color_palette("magma")
```

```python
datapath = "../Daten/"
zylFiles = []
for file in os.listdir(datapath+"Zyinder/neu"):
    if file.endswith(".dat"):
        print(file)
        zylFiles.append(os.path.join(datapath+"Zyinder/neu",file))
```

```python
for file in zylFiles:
    plt.clf()
    plt.close()
    fig = plt.figure()
    ax = fig.add_subplot()
    freq, amp = np.genfromtxt(file,unpack=True)
    ax.plot(freq/1000,amp,c=cmap[0])
    plt.xlabel(r"$Frequenz\:/\: kHz$")
    plt.ylabel(r"$Amplitude\:/\: a.u.$")
    plt.savefig(file.replace(".dat",".pdf"))
```

# Wasserstoffatom

```python
wassFiles = []
for file in os.listdir(datapath+"Wasserstoff/neu"):
    if file.endswith(".dat"):
        print(file)
        wassFiles.append(os.path.join(datapath+"Wasserstoff/neu",file))
print(len(wassFiles))
```

```python
from scipy.signal import find_peaks
```

```python
peakFreq = []
peaksat=[]
for file in wassFiles:
    plt.clf()
    plt.close()
    
    freq, amp = np.genfromtxt(file,unpack=True)
    
        
    plt.plot(freq/1000,amp,c=cmap[0])
    if "H_180.dat" in file:
        print(file)
        peaks,_ = find_peaks(amp, height=20, distance=5)
        
        plt.plot(freq[peaks]/1000,amp[peaks],'x',c=cmap[3])
        for i,peak in enumerate(peaks):
            if i < 6 and i != 2 and i != 4:
                peakFreq.append(freq[peak])
                peaksat.append(peak)
                text = "{}".format(np.around(freq[peak]/1000,2))
                plt.text(freq[peak]/1000-1,amp[peak]+1,r"f="+text+r" kHz",color='C1', fontsize='small',c=cmap[3])
    plt.xlabel(r"$Frequenz\:/\: kHz$")
    plt.ylabel(r"$Amplitude\:/\: a.u.$")
    plt.grid()
    if "H_180.dat" in file:
        plt.savefig(file.replace(".dat",".pdf"))
        plt.show()
        print(freq[peaksat])
    
    
```

```python
distances = [0]
for i,file in enumerate(spaltFile):
    plt.clf()
    plt.close()
    fig = plt.figure()
    ax = fig.add_subplot()
    freq, amp = np.genfromtxt(os.path.join(pathWass,file),unpack=True)
    if("H_0.dat" in file):
        ind = np.where((freq >1800) & (freq < 2600))
        freq = freq[ind]
        amp = amp[ind]
    peaks,_ = find_peaks(amp, height= 7)
    if("H_0.dat" not in file):
        dist = freq[peaks[1]]-freq[peaks[0]]
        print(dist)
        distances.append(dist)
        plt.text((freq[peaks[0]]+20)/1000,11,"{} Hz".format(int(dist)),c="r")
        plt.hlines(10,freq[peaks[0]]/1000, (freq[peaks[0]]+dist)/1000)
    ax.plot(freq/1000,amp,c=cmap[0])
    ax.plot(freq[peaks]/1000,amp[peaks],'x',c=cmap[4])
    plt.xlabel(r"$Frequenz\:/\: kHz$")
    plt.ylabel(r"$Amplitude\:/\: a.u.$")
    print(os.path.join(pathWass,"spalt{}.pdf".format(i)))
    plt.savefig(os.path.join(pathWass,"spalt{}.pdf".format(i)))
    plt.show()
```
