```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import os
import matplotlib.colors as col
```

# Erstens Vorbereitung

```python
sns.set()
cmap = sns.color_palette("magma")
```

```python
datapath = "../Daten/"
zylFiles = []
for file in os.listdir(datapath+"Zyinder"):
    if file.endswith(".dat"):
        print(file)
        zylFiles.append(os.path.join(datapath+"Zyinder",file))
```

```python
for file in zylFiles:
    plt.clf()
    plt.close()
    fig = plt.figure()
    ax = fig.add_subplot()
    freq, amp = np.genfromtxt(file,unpack=True)
    ax.plot(freq/1000,amp,c=cmap[0])
    plt.xlabel(r"$Frequenz\:/\: kHz$")
    plt.ylabel(r"$Amplitude\:/\: a.u.$")
    plt.savefig(file.replace(".dat",".pdf"))
```

# Wasserstoffatom

```python
wassFiles = []
for file in os.listdir(datapath+"Wasserstoff"):
    if file.endswith(".dat"):
        print(file)
        wassFiles.append(os.path.join(datapath+"Wasserstoff",file))
```

```python
from scipy.signal import find_peaks
```

```python
peakFreq = []
peaksat=[]
for file in wassFiles:
    plt.clf()
    plt.close()
    
    freq, amp = np.genfromtxt(file,unpack=True)
    
        
    plt.plot(freq/1000,amp,c=cmap[0])
    if "H_180.dat" in file:
        print(file)
        peaks,_ = find_peaks(amp, height=8)
        
        plt.plot(freq[peaks]/1000,amp[peaks],'x',c=cmap[3])
        for i,peak in enumerate(peaks):
            if i < 6 and i != 2 and i != 4:
                peakFreq.append(freq[peak])
                peaksat.append(peak)
                text = "{}".format(np.around(freq[peak]/1000,2))
                plt.text(freq[peak]/1000-1,amp[peak]+1,r"f="+text+r" kHz",color='C1', fontsize='small',c=cmap[3])
    plt.xlabel(r"$Frequenz\:/\: kHz$")
    plt.ylabel(r"$Amplitude\:/\: a.u.$")
    if "H_180.dat" in file:
        plt.savefig(file.replace(".dat",".pdf"))
        plt.show()
    
    
```

```python
peaksat
```

```python
filesWinkel = ["H_{}.dat".format(n) for n in range(0,181,10)]
print(filesWinkel)
peaklocs = []
peak1 = []
peak2 = []
peak3 = []
peak4 = []
for file in filesWinkel:
    filepath= os.path.join(datapath+"Wasserstoff",file)
    print(filepath)
    freq, amp = np.genfromtxt(filepath,unpack=True)
    
    peaks,_ = find_peaks(amp, height=9)
    freq = freq[peaksat]
    amp = amp[peaksat]
    peak1.append(amp[0])
    peak2.append(amp[1])
    peak3.append(amp[2])
    peak4.append(amp[3])
    peaklocs.append(freq)
```

```python
peakall = [peak1,peak2,peak3,peak4]
```

```python
from scipy.special import legendre
```

```python
theta = np.arange(0,181,10)*2*np.pi/360
phi = np.arccos(0.5*np.cos(theta) - 0.5)
phi2 = np.linspace(0,2*np.pi,100)
for i,p in enumerate(peakall):
    plt.clf()
    plt.close()
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    
    
    ax.plot(phi,p,'.-',c="C5")
    ax.plot(-phi,p,'.-',c="C5")
    ax.plot(np.pi+phi,p,'.-',c="C5")
    ax.plot(np.pi-phi,p,'.-',c="C5")
    plt.savefig(os.path.join("../Daten/Wasserstoff","peak{}.pdf".format(i)))
    plt.show()
    
```

```python
plt.rc('axes', unicode_minus=False)
from scipy.special import lpmv
```

```python
phi = np.linspace(0,2*np.pi,1000)
ns = [1,2,4,5]
for i,p in enumerate(peakall):
    plt.clf()
    plt.close()
    print(ns[i])
    pN = legendre(ns[i])
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.plot(phi,abs(pN(np.cos(phi))),c="C4")
    plt.yticks([])
    plt.savefig(os.path.join("../Daten/Wasserstoff","peakLeg{}.pdf".format(i)))
    plt.show()
```

# Aufspaltung des Peaks

```python
spaltFile = ["H_0.dat", "H_3mm.dat","H_6mm.dat","H_9mm_0.dat"]
```

```python
pathWass = "../Daten/Wasserstoff"
```

```python
distances = [0]
for i,file in enumerate(spaltFile):
    plt.clf()
    plt.close()
    fig = plt.figure()
    ax = fig.add_subplot()
    freq, amp = np.genfromtxt(os.path.join(pathWass,file),unpack=True)
    if("H_0.dat" in file):
        ind = np.where((freq >1800) & (freq < 2600))
        freq = freq[ind]
        amp = amp[ind]
    peaks,_ = find_peaks(amp, height= 7)
    if("H_0.dat" not in file):
        dist = freq[peaks[1]]-freq[peaks[0]]
        print(dist)
        distances.append(dist)
        plt.text((freq[peaks[0]]+20)/1000,11,"{} Hz".format(int(dist)),c="r")
        plt.hlines(10,freq[peaks[0]]/1000, (freq[peaks[0]]+dist)/1000)
    ax.plot(freq/1000,amp,c=cmap[0])
    ax.plot(freq[peaks]/1000,amp[peaks],'x',c=cmap[4])
    plt.xlabel(r"$Frequenz\:/\: kHz$")
    plt.ylabel(r"$Amplitude\:/\: a.u.$")
    print(os.path.join(pathWass,"spalt{}.pdf".format(i)))
    plt.savefig(os.path.join(pathWass,"spalt{}.pdf".format(i)))
    plt.show()
```

```python
import pandas as pd
```

```python
rings = [0,3,6,9]
d = {'ring': rings, 'ab': distances}
df = pd.DataFrame(data=d)
plt.clf()
plt.close()
sns.regplot(x="ring",y= "ab", data=df,color=cmap[4])
plt.xlabel(f"Ring / $mm$")
plt.ylabel(f"Abstand / $Hz$")
plt.xlim(-0.4,9.4)
plt.tight_layout()
plt.savefig(os.path.join(pathWass,"spaltGesamt.pdf"))
plt.show()
```

# Druckamplitude bei 9 mm. Wasserstoff

```python
plt.clf()
plt.close()
fig = plt.figure()
ax = fig.add_subplot()
freq, amp = np.genfromtxt(os.path.join(pathWass,"H_9mm_0.dat"),unpack=True)
peaks,_ = find_peaks(amp, height= 7)
ax.plot(freq/1000,amp,c=cmap[0])
ax.plot(freq[peaks[1]]/1000,amp[peaks[1]],'x',c="red")
plt.xlabel(r"$Frequenz\:/\: kHz$")
plt.ylabel(r"$Amplitude\:/\: a.u.$")
peakfreq = int(freq[peaks[1]])
print("In diesem Abschnitt wird die Resonanz bei " + str(int(peakfreq))+" untersucht.")
plt.show()
```

```python
filesWinkel = ["H_9mm_{}.dat".format(n) for n in range(0,181,10)]
print(filesWinkel)
peak = []
for file in filesWinkel:
    filepath= os.path.join(pathWass,file)
    print(filepath)
    freq, amp = np.genfromtxt(filepath,unpack=True)
    peak.append(float(amp[np.where(freq==peakfreq)]))
```

```python
theta = np.arange(0,181,10)*2*np.pi/360
phi = np.arccos(0.5*np.cos(theta) - 0.5)


plt.clf()
plt.close()
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})


ax.plot(phi,peak,'.-',c="C5")
ax.plot(-phi,peak,'.-',c="C5")
ax.plot(np.pi+phi,peak,'.-',c="C5")
ax.plot(np.pi-phi,peak,'.-',c="C5")
plt.savefig(os.path.join("../Daten/Wasserstoff","peak9mm.pdf"))
plt.show()
```

```python
phi = np.linspace(0,2*np.pi,1000)
plt.clf()
plt.close()
pN = legendre(2)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(phi,abs(pN(np.cos(phi))),c="C4")
plt.yticks([])
plt.show()
```

# Wasserstoffmolekül

```python
datapath = "../Daten/"
molFiles = []
for file in os.listdir(datapath+"Wasserstoffmolekuelion"):
    if file.endswith(".dat"):
        print(os.path.join(datapath+"Wasserstoffmolekuelion",file))
        molFiles.append(os.path.join(datapath+"Wasserstoffmolekuelion",file))
```

```python
for file in molFiles:
    plt.clf()
    plt.close()
    fig = plt.figure()
    ax = fig.add_subplot()
    freq, amp = np.genfromtxt(file,unpack=True)
    ax.plot(freq/1000,amp,c=cmap[0])
    plt.xlabel(r"$Frequenz\:/\: kHz$")
    plt.ylabel(r"$Amplitude\:/\: a.u.$")
    plt.savefig(file.replace(".dat",".pdf"))
```

```python
bH2 = ["H2_5mm.dat","H2_10mm.dat","H2_15mm.dat","H2_20mm_180.dat"]
d = [5,10,15,20]
plt.clf()
plt.close()
fig = plt.figure()
ax = fig.add_subplot()
path = "../Daten/Wasserstoffmolekuelion"
for i,file in enumerate(bH2):  
    freq, amp = np.genfromtxt(os.path.join(path,file),unpack=True)
    ind = np.where(freq>2300)
    ind2 = np.where(freq<=2300)
    ax.plot(freq[ind2]/1000,amp[ind2],':',c=cmap[i])
    ax.plot(freq[ind]/1000,amp[ind],c=cmap[i],label=r"$d = {}\, mm$".format(d[i]))
    
    peaks,_ = find_peaks(amp[ind], distance=2000)
    ax.plot(freq[ind][peaks]/1000,amp[ind][peaks],'x',c="red")
    plt.text(freq[ind][peaks]/1000-0.02,amp[ind][peaks]+0.3, r"${}\, Hz$".format(int(freq[ind][peaks])),c=cmap[3],fontsize=10)
    
plt.xlabel(r"$Frequenz\:/\: kHz$")
plt.ylabel(r"$Amplitude\:/\: a.u.$")
plt.tight_layout()
plt.legend(loc="best")
plt.savefig("../Daten/Wasserstoffmolekuelion/blendAbhaeng.pdf")
plt.show()
```

```python
peaks1=[]
peaks2=[]
for i,file in enumerate(bH2):
    print(os.path.join(path,file))
    plt.clf()
    plt.close()
    fig = plt.figure()
    ax = fig.add_subplot()
    freq, amp = np.genfromtxt(os.path.join(path,file),unpack=True)
    ind = np.where((freq<2290)& (freq>2260))
    freq=freq[ind]
    amp =amp[ind]
    peaks,_ = find_peaks(amp, height=4, distance=50)
    peaks1.append(freq[peaks[0]])
    peaks2.append(freq[peaks[1]])
    ax.plot(freq/1000,amp,c=cmap[0])
    ax.plot(freq[peaks]/1000,amp[peaks],'x',c=cmap[4])
    plt.xlabel(r"$Frequenz\:/\: kHz$")
    plt.ylabel(r"$Amplitude\:/\: a.u.$")
    plt.show()
```

```python
peaks3 = [2311,2355,2398,2437]
print(peaks1)
print(peaks2)
```

```python
plt.clf()
plt.close()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(d,peaks1,'o-',c=cmap[0],label="Peak 1")
ax.plot(d,peaks2,'o-',c=cmap[2],label="Peak 2")
ax.plot(d,peaks3,'o-',c=cmap[4],label="Peak 3")
plt.xlabel(r"$Blendendurchmesser\:/\:mm$")
plt.ylabel(r"$Resonanz\:/\: kHz$")
plt.show()
```
